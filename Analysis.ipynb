```python
import pandas as pd

user_names = ["UserID", "Gender", "Age", "Occupation", "Zip-code"]
rnames = ["UserID", "MovieID", "Rating", "Timestamp"]
mnames = ["MovieID", "Title", "Genres"]

users = pd.read_table(r"C:\Users\ronob\Desktop\Movie dataset analysis\ml-1m\ml-1m\users.dat", sep='::', header=None, names=user_names, engine='python')
ratings = pd.read_table(r"C:\Users\ronob\Desktop\Movie dataset analysis\ml-1m\ml-1m\ratings.dat", sep='::', header=None, names=rnames, engine='python')
#movies = pd.read_table(r"C:\Users\ronob\Desktop\Movie dataset analysis\ml-1m\ml-1m\movies.dat", sep='::', header=None, names=mnames, engine='python')

with open(r"C:\Users\ronob\Desktop\Movie dataset analysis\ml-1m\ml-1m\movies.dat", 'rb') as f:
    result = chardet.detect(f.read())

movies = pd.read_table(r"C:\Users\ronob\Desktop\Movie dataset analysis\ml-1m\ml-1m\movies.dat", sep='::', header=None, names=mnames, engine='python', encoding=result['encoding'])
```

```python
data=pd.merge(pd.merge(ratings,users),movies)
data
```

```python
# Mean movie rating grouped by gender
mean_rating= pd.pivot_table(data, index='Title', columns='Gender', aggfunc='mean')
mean_rating
```

```python
#data frame with movie title as row lables ad ratings as columns
rating_by_title=data.groupby('Title').size()
rating_by_title
#filtering movies with atleast 250 rating
Active_title= rating_by_title.index[rating_by_title>250]
Active_title
```

```python
mean_rating= mean_rating.loc[Active_title]
mean_rating
```

```python
top_female_ratings = mean_rating[mean_rating['Gender'] == 'F'].sort_values(by='mean_rating_column', ascending=False)
```
